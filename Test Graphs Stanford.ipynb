```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

import os
import h5py
import scipy.io

BATCH = 2

FN=f"batch{BATCH}_mit.mat"
arrays = {}
mat = h5py.File(FN)





def create_df(cell, features=["t", "I", "V", "T"], K = 10):
    k = mat["batch"]["cycles"][cell][0]
    obj = mat[k]
    df = {i:[] for i in features}
    df["cycle"] = []
    for i in range(1, len(obj["I"])):
        for j in features:
            df[j].extend(list(obj[obj[j][i,0]][0])) 
        N = len(obj[obj["I"][i,0]][0])
        df["cycle"].extend([i]*N)
    df =  pd.DataFrame(df)
    
    df = df.loc[df.cycle>=K]
    M = max(max(df.cycle)-K, K+2)
    df = df.loc[df.cycle<=M]
    return df

df = create_df(1)
epsilon = 0.05
cycle_step = 1





def clear_outliers(L, K = 3, start = 10):
    n = len(L)
    for i in range(start, n):
        V = sum(L[i-K:i])/K
        if abs(V)>0 and abs(L[i]-V)/abs(V)>1:
            L[i] = V
```
